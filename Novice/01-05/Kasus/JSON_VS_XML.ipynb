# XML

```python
import xml.etree.ElementTree as ET
tree = ET.parse('resume.xml')
root = tree.getroot()
print(root)
```

```python
for child in root:
    print(child.tag, child.attrib)
```

```python
[elem.tag for elem in root.iter()]
```

```python
print(ET.tostring(root, encoding='utf8').decode('utf8'))
```

XML lebih susah diurai (parse) kalimatnya daripada JSON. 
JSON lebih cepat daripada XML karena library DOM membutuhkan memory yang lebih besar untuk memproses file ukuran besar di XML. 
JSON menggunakan lebih sedikit data, sehingga mengurangi cost dan meningkatkan kecepatan parsing.
JSON lebih strutktur dan mudah dibaca.
